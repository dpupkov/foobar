```python
def solution_foo(start, length):
    start = int(start)
    length = int(length)
    if start > 2000000000:
        return(0)
    if length < 1:
        return(0)
    result = 0
    
    for s in range(length):
        for worker in range(start + s*length, start + (s+1)*length - s):
            if worker > 2000000000:
                result = result ^ (worker - 2000000001)
            else:
                result = result ^ worker
#        print('result: ', result)
    return result
```

```python
def xor(x):
    if x % 4 == 0:
        return(x)
    elif x % 4 == 1:
        return(1)
    elif x % 4 == 2:
        return(x+1)
    elif x % 4 == 3:
        return(0)
```

```python
def solution(start, length):
    result = 0
    for s in range(length):
        s1 = start + s*length
        s2 = start + (s+1)*length - s - 1
#        print("s, s1, s2: ",s, s1, s2)
#        print('before: ', result)
        if s1 == s2:
            result ^= s1
        else:
            result ^= xor(s1 - 1) ^ xor(s2)
                
#        print('result: ', result)
#             if worker > 2000000000:
#                 result = result^(worker - 2000000001)
#             else:
#                 result = result^worker
    return result
```
